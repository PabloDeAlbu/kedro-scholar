# URLs de IR en OpenAIRE

Los handle de los recursos en OpenAIRE se encuentran en el campo url. Este campo contiene la URL del recurso recuperado por OpenAIRE.

Es posible obtener los handle filtrando los mismo con el prefijo del handle institucional. El repositorio de la UNLP tiene asignado el handle _10915_

```python
handle_filter = dim_url_openaire["url"].str.contains("handle/10915", na=False)
dim_handle_openaire = dim_url_openaire[handle_filter]

publication_handle_openaire = pd.merge(
    bridge_publication_url_openaire,
    dim_handle_openaire
)
publication_handle_openaire.drop(columns=['url_hk'], inplace=True)
publication_handle_openaire["handle"] = publication_handle_openaire["url"].str.extract(r"(10915/\d+)")
```
