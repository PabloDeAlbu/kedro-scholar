```python
def fetch_openalex_institution(institution_ror, env):
    cursor = '*'
    base_url = 'https://api.openalex.org/institutions/ror:{}'

    url = base_url.format(institution_ror, cursor)
    api_response = requests.get(url).json()
    return api_response
    
    keys_to_delete = ['lineage','display_name_acronyms','display_name_alternatives','image_url','image_thumbnail_url','summary_stats','geo','international','associated_institutions','counts_by_year','roles','topics','topic_share','x_concepts','is_super_system','works_api_url']
    institution_dict = {key: value for key, value in api_response.items() if key not in keys_to_delete}
    df = pd.DataFrame.from_dict(institution_dict)

    df['load_datetime'] = date.today()

    return df
```
