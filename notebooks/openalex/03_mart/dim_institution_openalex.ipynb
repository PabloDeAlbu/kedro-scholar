```python
def dim_institution_openalex(sal_dinstitution_openalex, sat_dinstitution_openalex): 
    
    dim_institution = pd.merge(
        sal_dinstitution_openalex, sat_dinstitution_openalex, 
        on="dinstitution_hk", 
        suffixes=('', '_remove')
    )

    columns_to_delete = dim_institution.filter(regex='(_remove$|_hk$|^hashdiff$|^source$|^load_datetime$)').columns
    dim_institution.drop(columns=columns_to_delete, inplace=True)
    
    return dim_institution
```
