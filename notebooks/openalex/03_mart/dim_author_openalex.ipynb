```python
def dim_author_openalex(sal_author_openalex, sat_author_openalex): 
    
    dim_author = pd.merge(
        sal_author_openalex, sat_author_openalex, 
        on="dauthor_hk", 
        suffixes=('', '_remove')
    )

    columns_to_delete = dim_author.filter(regex='(_remove$|^hashdiff$|^source$|^load_datetime$)').columns
    dim_author.drop(columns=columns_to_delete, inplace=True)
    
    return dim_author
```
