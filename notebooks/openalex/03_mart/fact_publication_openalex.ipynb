## Creación de **fact_publication_openalex** a partir de **sal_publication_openalex** y **sat_publication_openalex**

**sal_publication_openalex** es un tabla link _same_as_link_ que reúne todos los identificadores persistentes de el repositorio institucional

**sat_publication_openalex** es el satélite con los datos descriptivos para una publicación en OpenAlex 

```python
def fact_publication_openalex(sal_work_openalex, sat_work_openalex): 
    fact_publication = pd.merge(
        sal_work_openalex, sat_work_openalex, 
        on="work_hk", 
        suffixes=('', '_remove')
    )
    columns_to_delete = fact_publication.filter(regex='(_remove$|^hashdiff$|^source$|^load_datetime$)').columns
    fact_publication.drop(columns=columns_to_delete, inplace=True)
    return fact_publication
```

```python
fact_publication = fact_publication_openalex(sal_work_openalex, sat_work_openalex)
```
