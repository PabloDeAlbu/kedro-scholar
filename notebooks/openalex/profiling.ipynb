Se utiliza el ROR para recuperar

```python
institution_ror
```

```python
# Merge entre sal_institution y hub_ror (Outer para no perder instituciones sin ROR)
dim_institution = pd.merge(
    sal_institution, hub_ror, 
    on="ror_hk"
)

dim_institution = pd.merge(dim_institution, sat_institution, on='institution_hk', )
dim_institution = dim_institution[['institution_hk','display_name','ror_x','country_code']].rename(columns = {'ror_x': 'ror'})
dim_institution
```

```python
ror_filter = dim_institution['ror'] == institution_ror
dim_institution[ror_filter]
```

```python
# filtro autores institucionales
institution_hk = dim_institution[ror_filter]['institution_hk'].iloc[0]

link_author_institution = link_author_institution[link_author_institution['institution_hk'] == institution_hk]
institutional_authors = pd.merge(link_author_institution, sat_author, on='author_hk')
institutional_authors = institutional_authors[['author_hk','author_id', 'orcid', 'display_name', 'works_count', 'cited_by_count']]

institutional_authors[['author_id', 'orcid', 'display_name', 'works_count', 'cited_by_count']]
```

Recupero trabajos de los autores institucionales en OpenAlex

```python
bridge_work_author = pd.merge(link_work_author, institutional_authors, on='author_hk')
```

```python
institutional_works = pd.merge(bridge_work_author, hub_work, on='work_hk')[['work_hk','work_id']]
fact_publication = institutional_works.drop_duplicates()
fact_publication
```
