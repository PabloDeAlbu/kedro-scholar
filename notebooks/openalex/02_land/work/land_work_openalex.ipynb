```python
def land_work_openalex(df_work_raw):
    df_work = df_work_raw.loc[:,
        [
            'id',
            'doi',
            'title',
            'display_name',
            'publication_year',
            'publication_date',
            'language',
            'type',
            'type_crossref',
            'countries_distinct_count',
            'institutions_distinct_count',
            'fwci',
            'has_fulltext',
            'fulltext_origin',
            'cited_by_count',
            'is_retracted',
            'is_paratext',
            'locations_count',
            'referenced_works_count',
            'cited_by_api_url',
            'updated_date',
            'created_date'
        ]
    ]
    df_work = df_work.convert_dtypes()
    df_work.rename(columns={'id':'work_id'}, inplace=True)

    
    return df_work
```
