```python
def land_work2sustainable_development_goals_openalex(df_work_raw):
    df_work = df_work_raw.loc[:,['id','sustainable_development_goals']]
    df_work = df_work.convert_dtypes()
    
    df_exploded =  df_work.explode('sustainable_development_goals')
    df_norm = pd.json_normalize(df_exploded['sustainable_development_goals'])
    df_work2sustainable_development_goals = pd.concat((df_work_raw['id'].reset_index(drop=True), df_norm.reset_index(drop=True)), axis=True)
    return df_work2sustainable_development_goals
```
